#Image Compression and Decompression using PCA

```python
# Import Libraries
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

#Load the Image convert
image= Image.open('lena_gray.gif').convert("L")
image_array = np.array(image, dtype=np.float64) # Convert to NumPy array (float for PCA calculations)

#Center the Dataset (Mean Subtraction)
mean= np.mean(image_array, axis=0)
centered_data = image_array - mean

#Standardize the Data (Optional but improves PCA)
std_dev= np.std(centered_data, axis=0)
std_dev[std_dev == 0] = 1 #Avoid division by zero
standardized_data = centered_data / std_dev

#Compute Covariance Matrix
cov_matrix = np.cov(standardized_data, rowvar=False) # Covariance matrix

#Display Original Image
plt.imshow(image_array, cmap="gray")
plt.title("Original Grayscale Image")
plt.axis("off")
plt.show()

#Print key statistics
print("Image Shape:", image_array.shape)
print("Mean Shape:", mean.shape)
print("Covariance Matrix Shape:", cov_matrix.shape)
```

2. Eigen Decomposition and Identifying Principal Components:
 Compute Eigen Values and Eigen Vectors.
 Sort the eigenvalues in descending order and choose the top k eigenvectors corresponding to
the highest eigenvalues.
 Identify the Principal Components with the help of cumulative Sum plot.

```python
# Step 1: Compute Eigenvalues and Eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

# Step 2: Sort Eigenvalues in Descending Order
sorted_indices = np.argsort(eigenvalues)[::-1]  # Indices of sorted eigenvalues
eigenvalues = eigenvalues[sorted_indices]
eigenvectors = eigenvectors[:, sorted_indices]  # Arrange eigenvectors in descending order

# Step 3: Compute Cumulative Sum of Explained Variance
explained_variance = eigenvalues / np.sum(eigenvalues)  # Explained variance ratio
cumulative_variance = np.cumsum(explained_variance)  # Cumulative sum

# Step 4: Plot the Explained Variance
plt.figure(figsize=(8,5))
plt.plot(cumulative_variance, linestyle='-', color='b')
plt.xlabel("Number of Principal Components")
plt.ylabel("Cumulative Explained Variance")
plt.title("Explained Variance vs. Number of Principal Components")
plt.grid()
plt.show()
```

Reconstruction and Experiment:

 Reconstruction: Transform the original data by multiplying it with the selected eigenvec-
tors(PCs) to obtain a lower-dimensional representation.

 Experiments: Pick Four different combination of principal components with various ex-
plained variance value and compare the result.

 Display the Results and Evaluate.

```python
def reconstruct_image(k):
    """Reconstructs image using top k principal components"""
    top_k_eigenvectors = eigenvectors[:, :k]  # Select top-k eigenvectors
    lower_dimensional_representation = np.dot(standardized_data, top_k_eigenvectors)  # Project to lower dimension
    reconstructed_data = np.dot(lower_dimensional_representation, top_k_eigenvectors.T)  # Reconstruct back
    reconstructed_data = (reconstructed_data * std_dev) + mean  # Reverse standardization
    return np.clip(reconstructed_data, 0, 255).astype(np.uint8)  # Clip values & convert to 8-bit integer

# Experiment with different k values
k_values = [5, 20, 50, 100]  # Different levels of compression
fig, axes = plt.subplots(1, 5, figsize=(15, 5))

# Display Original Image
axes[0].imshow(image_array, cmap="gray")
axes[0].set_title("Original Image")
axes[0].axis("off")

# Display Reconstructed Images with Different k
for i, k in enumerate(k_values):
    reconstructed_image = reconstruct_image(k)
    axes[i + 1].imshow(reconstructed_image, cmap="gray")
    axes[i + 1].set_title(f"k = {k}")
    axes[i + 1].axis("off")

plt.show()
```
